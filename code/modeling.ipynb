```python
def split_features(feat, target_col):
    X_train = feat[feat.index < SPLIT_DATE].drop(columns=[target_col])
    X_test  = feat[feat.index >= SPLIT_DATE].drop(columns=[target_col])
    y_train = feat[feat.index < SPLIT_DATE][target_col]
    y_test  = feat[feat.index >= SPLIT_DATE][target_col]
    print(f'Train: {X_train.shape}  Test: {X_test.shape}')
    return X_train, X_test, y_train, y_test
```
